In this exercise, you will be asked to explore time series data in several ways.

You will be working with time series of Open Power System Data ([OPSD](https://open-power-system-data.org/)) for Germany. 

The data set includes daily electricity consumption, wind power production, and solar power production between 2006 and 2017. 
```
- Date — The date (yyyy-mm-dd format)
- Consumption — Electricity consumption in GWh
- Wind — Wind power production in GWh
- Solar — Solar power production in GWh
- Wind+Solar — Sum of wind and solar power production in GWh
```

### 1. Plotting and understanding a time series

Q1 - Plot the different time series. What do you understand, what do you notice ?

````python
plt.figure(figsize=(12,8))
plt.plot( df["Consumption"],color = 'blue',linewidth = 0.2, label = "Consumption")
plt.plot(df["Wind+Solar"],color = 'orange',linewidth = 0.5,label = "Production")
plt.xlabel("Year")
plt.ylabel("GWh")
plt.legend()
plt.show()
````

It can be seen from the chart that there is a higher level of comsumption than production, although seems like there is a trend of production being increased over time. Also, the comsumption has a seasonal pattern, with low and high seasons.

We can also avoid plotting lines and simply plot dots for the consumption :

````python
plt.figure(figsize=(12,8))
plt.plot(df["Consumption"],color = 'blue',linestyle = "None", marker = ".",label = "Consumption")
plt.xlabel("Year")
plt.ylabel("GWh")
plt.legend()
plt.show()
````

And plot the production over time :

````python
plt.figure(figsize=(12,8))
plt.plot( df["Wind"],color = 'blue',linestyle = "None", marker = ".",label = "Wind")
plt.plot( df["Solar"],color = 'orange',label = "Solar")
plt.legend()
plt.show()
````

What can we notice ?
+ the consumption is larger than the production
+ we have no overall production data before 2012
+ the production and the consumption are both seasonal, i.e. they vary within a year given some pattern
+ there is a trend in production (the higher part), and outliers downwards

How do consumption and production compare ?
+ We notice that wind bsaed production increases over time, and is rather high in winter (for example more storms).
+ On the other hand, the solar production is higher in summer (seems logic)
+ The consumption is higher in winter

````python
plt.figure(figsize=(12,8))
plt.plot(df.loc["2012-01-01":"2018-01-01"]["Consumption"],color = 'blue',linewidth = 0.2, label = "Consumption")
plt.plot(df.loc["2012-01-01":"2018-01-01"]["Wind"],color = 'orange',linewidth = 0.5,label = "Wind")
plt.plot(df.loc["2012-01-01":"2018-01-01"]["Solar"],color = 'green',linewidth = 0.5,label = "Solar")
plt.xlabel("Year")
plt.ylabel("GWh")
plt.legend()
plt.show()
````

### 2. Change the scale

Q2 - Try to change the scale to understand the recurrent patterns !

````python
# Plot over the year
plt.figure(figsize=(12,8))
plt.plot(df.loc["2017-01-01":"2018-01-01"]["Consumption"],color = 'blue',linewidth = 0.2, label = "Consumption")
plt.title("Electricity Comsumption in 2017")
plt.show()
````

It confirms the trend : more consumption in winter time, and less in summer. We can then zoom on a given month :

````python
# Plot over a month
plt.figure(figsize=(12,8))
plt.plot(df.loc["2017-07-01":"2017-08-01"]["Consumption"],color = 'blue',linewidth = 0.2, label = "Consumption")
plt.title("Electricity Comsumption in July 2017")
plt.show()
````

When zooming on a month, we oberve another seasonality. There seems to be days of the week in which people consume less. For example, here's what happens when we highlight Sundays :

````python
# Plot on Sundays
plt.figure(figsize=(12,8))

plt.plot(df.loc["2017-07-01":"2017-08-01"]["Consumption"],color = 'blue',linewidth = 0.2, label = "Consumption")
plt.axvline(pd.to_datetime("2017-07-02"),color = 'red')
plt.axvline(pd.to_datetime("2017-07-09"),color = 'red')
plt.axvline(pd.to_datetime("2017-07-16"),color = 'red')
plt.axvline(pd.to_datetime("2017-07-23"),color = 'red')
plt.axvline(pd.to_datetime("2017-07-30"),color = 'red')
plt.title("Electricity Comsumption in July 2017")
plt.show()
````

### 3. Further exploration

Q3 - To further explore the dataset, we might need to group the data. Create new columns for the consumption (by year and month), and create box-plots with seaborn. Comment on your results.

````python
# new columns 'year' and 'month'
df_2 = df.reset_index()
df_2['year'] = [d.year for d in df_2.Date]
df_2['month'] = [d.strftime("%b") for d in df_2.Date]
years = df_2['year'].unique()
print(years)
df_2.head()
````

````python
# Boxplot year
import seaborn as sns
plt.figure(figsize=(12,8))
sns.boxplot(x="year",y="Consumption", data = df_2)
plt.show()
````

Apart from the year 2009 the consumption hasn't been changed much along the years. It is possible to see that there are outliers in the data, in which the comsumption varies highly.

````python
# Boxplot month
plt.figure(figsize=(12,8))
plt.figure(figsize=(12,8))
sns.boxplot(x="month",y="Consumption", data = df_2)
plt.show()
````

The seasonality across the months are visible in the chart, along the months of may to august. Also there are high variance in the data across the month.
